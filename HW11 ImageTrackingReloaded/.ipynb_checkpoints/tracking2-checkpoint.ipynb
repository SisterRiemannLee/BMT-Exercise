```python
%matplotlib notebook
import scipy.io as sio
import scipy.ndimage as ndimage
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
```

```python
def golub(a, b):
    # QR decomposition of A
    [q, r] = np.linalg.qr(a)

    # compute new b
    bnew = np.matmul(q.T, b)

    # get size of A
    [m, n] = a.shape
    # compute solution
    x = np.linalg.solve(r[0:n, :], bnew[0:n])

    # compute residual
    residual = np.sqrt(np.sum((np.matmul(a, x) - b)**2))

    return x, residual
```


## load video and set up coordinate system


```python
mat_contents = sio.loadmat('tracking-cars.mat')
# mat_contents = sio.loadmat('tracking-leukocyte.mat')
```

### convert movie to double


```python
movie = np.asarray(mat_contents['movie'], np.double)
```


### get first image (movie is a 3-d array, where the 3rd dimension is the frame number)


```python
Is = movie[:, :, 0]
```

### generate coordinate system


```python
[r, c] = Is.shape
x = list(range(0, c))
y = list(range(0, r))
```


### display image


```python
plt.imshow(Is, cmap='gray')
```


### get center point for patch P


```python
# for testing you can use these coordinates
tr_x = [180-1]
tr_y = [147-1]

#tr = plt.ginput(1)

#tr_x = [tr[0][0]]
#tr_y = [tr[0][1]]
```


### track the patch


```python
# define 'radius' size of patch
t = 10

# set up patch coordinate system
# set up patch coordinate system
[tx, ty] = np.meshgrid(range(-t, t+1), range(-t, t+1))
tx = tx.flatten()
ty = ty.flatten()

# display image again
plt.imshow(Is, cmap='gray')

# initialize other trajectories
tr_x_basic = list(tr_x)
tr_y_basic = list(tr_y)


# draw trajectory
plt.plot(tr_x, tr_y, 'r.')
plt.plot(tr_x_basic, tr_y_basic, 'g.')

# tell pyplot that we want an asynchronous figure
plt.ion()
```

```python
# for later drawing the tracking result, we store in "lines" the sequence of lines, i.e. a list of the growing lines
lines = []

# loop over all images
for i in range(1, movie.shape[2]):
    # get target image
    It =  movie[:, :, i]
    
    # compute function and derivatives
    # compute gradient of source image, the result are two 'images'
    # storing d_xI_s and d_yI_s
    [dyIs, dxIs] = np.gradient(Is)

    # compute update using newton method
    v = [0, 0]
    v_basic = [0, 0]

    for j in range(0,15):
        # note on interpolation:
        # [tr_x[-1]]   [ tx ]   [ v[0] ]
        # [        ] + [    ] + [      ]
        # [tr_y[-1]]   [ ty ]   [ v[1] ]
        # is the trajectory's end evaluated at all positions of the
        # patch shifted by v

        # evaluate g
        g = ndimage.map_coordinates(Is.T, (tr_x[-1] + tx + v[0], tr_y[-1] + ty + v[1]), order=2) - \
            ndimage.map_coordinates(It.T, (tr_x[-1] + tx, tr_y[-1] + ty), order=2)
        g = g.flatten()

        # evaluate J_g
        J_g1 = ndimage.map_coordinates(dxIs.T, (tr_x[-1] + tx + v[0], tr_y[-1] + ty + v[1]), order=2)
        J_g2 = ndimage.map_coordinates(dyIs.T, (tr_x[-1] + tx + v[0], tr_y[-1] + ty + v[1]), order=2)
        J_g = np.stack((J_g1.T.flatten(), J_g2.T.flatten()), axis=-1)

        # compute update
        updateX, reisudual = golub(J_g, g)
        update = -updateX

        # perform Newton step
        v = v + update

    # compute basic method based on taylor expansion

    # get all pixels of patch in d_xI_s
    # and store it in a variable called A1.
    # Note that we have to take the current endpoint 
    # of the trajectory (tr_x[-1],tr_y[-1]).
    A1 = ndimage.map_coordinates(dxIs.T, (tr_x_basic[-1] + tx, tr_y_basic[-1] + ty), order=2)

    # get all pixels of patch in d_yI_s
    # and store it in a variable called A2.
    A2 = ndimage.map_coordinates(dyIs.T, (tr_x_basic[-1] + tx, tr_y_basic[-1] + ty), order=2)
    
    # get all pixels of patch in I_t - I_s 
    # and store it in a variable called b.
    b = ndimage.map_coordinates(It.T, (tr_x_basic[-1] + tx, tr_y_basic[-1] + ty), order=2) - \
        ndimage.map_coordinates(Is.T, (tr_x_basic[-1] + tx, tr_y_basic[-1] + ty), order=2)
    
    # set up equation system
    # for that transform A1, A2 and b to column vectors
    A = np.stack((A1.T.flatten(), A2.T.flatten()), axis=-1)
    b = b.flatten()
    
    # solve equation system using Golub's algorithm
    [v_basic, res] = golub(A, b)
    
    # update trajectory
    # Attention! I_s(x+v) ~ I_t(x)
    # since we show always It, we have to subtract v !
    tr_x.append(tr_x[-1] - v[0])
    tr_y.append(tr_y[-1] - v[1])
    tr_x_basic.append(tr_x_basic[-1] - v_basic[0])
    tr_y_basic.append(tr_y_basic[-1] - v_basic[1])

    lines.append((np.asarray(tr_x), np.asarray(tr_y)))

    # update Is
    Is = It
```


### plot tracking output


```python
fig = plt.figure()
ax = plt.axes()

line, = ax.plot([], [], lw=1)
img = ax.imshow(movie[:,:,0], cmap='gray')

# Initialization function: plot the background of each frame
def init():
    line.set_data(lines[0][0], lines[0][1])
    img.set_data(movie[:,:,0])
    return line,

# Animation function which updates figure data.  This is called sequentially
def animate(i):
    line.set_data(lines[i][0], lines[i][1])
    img.set_data(movie[:,:,i])
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=len(lines), interval=50)

# Call function to display the animation
_ = HTML(anim.to_html5_video())
```
