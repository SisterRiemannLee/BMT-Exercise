```python
# Lecture "Basic Math Tools"
# Supplementary Material
#
# Technische Universität München, Fakultät für Informatik
# Dr. Tobias Lasser, Richard Brosig, Jakob Vogel, Dr. Maximilian Baust, Rüdiger Göbl
# 2013

%matplotlib inline

from scipy import signal
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

width = 14
height = 10

# FUNCTION convolution_sl_v2(filter, boundary = 'trim')
#
# Demonstrates convolution, the parameters define the filter and control
# the way the boundary is handled. Possible values are:
#
# 'trim'        Discards values where the full data set is not available
# 'zero'        Pads with zero values.
def convolution_sl_v2(I, filt, boundary='zero'):

    # choose an appropriate parameter for 'convolve2d'
    param_python = 'same'
    if boundary == 'trim':
        param_python = 'valid'

    # compute the result using matlab's built-in function
    result_python = signal.convolve2d(I, filt, param_python)
    plt.figure(figsize=(width, height))
    plt.imshow(result_python, cmap='gray')
    plt.title('python (' + param_python + ')')

    # compute the result using our function
    result_our = do_convolution(I, filt, boundary)
    plt.figure(figsize=(width, height))
    plt.imshow(result_our, cmap='gray')
    plt.title('ours (' + boundary + ')')

    # compate the results using SSD
    diff = result_python - result_our
    plt.figure(figsize=(width, height))
    plt.imshow(diff)
    plt.title('difference')

    plt.figure(figsize=(width, height))
    plt.imshow(I, cmap='gray')
    plt.title('original')

    print('SSD = ' + str(np.sum(diff**2)))
    print('min(diff) = ' + str(np.min(diff)))
    print('max(diff) = ' + str(np.max(diff)))


def do_convolution(image, filt, boundary):
    # obtain the size of the image
    image_size = image.shape
    filter_size = filt.shape
    filter_origin = (filter_size[0] // 2, filter_size[1] // 2)

    # initialize the result with
    # NOTE: in-place computation of the convolution will not work!
    result = np.zeros(image_size)

    # run through the image in a naive implementation
    for i1 in range(0, image_size[0]):
        for i2 in range(0, image_size[1]):
            accumulated = 0
            for j1 in range(0, filter_size[0]):
                for j2 in range(0, filter_size[1]):
                    k1 = j1 - filter_origin[0]
                    k2 = j2 - filter_origin[1]
                    if i1 - k1 < 0 or i1 - k1 >= image_size[0] or i2 - k2 < 0 or i2 - k2 >= image_size[1]:
                        image_value = 0
                    else:
                        image_value = image[i1 - k1, i2 - k2]
                    accumulated = accumulated + filt[j1, j2] * image_value
            result[i1, i2] = accumulated

    # trim away problematic values
    if boundary == 'trim':
        result = result[1:-1, 1:-1]
    return result


# load a grayscale image
I = np.array(Image.open("Wikimedia_6538426675.png")) / 255
```

```python
# try the filters
myFilter = (1/9) * np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])
convolution_sl_v2(I, myFilter)
plt.show()
```

```python
myFilter = np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]]) / 16
convolution_sl_v2(I, myFilter)
plt.show()
```

```python
from scipy import ndimage

# number of scales
N = 4

# initial gamma
gamma = 0.5

# load image
I = np.array(Image.open("Wikimedia_6538426675.png"))/255

# define Gaussian filter
dirac = np.zeros([9, 9])
dirac[4, 4] = 1
G = ndimage.gaussian_filter(dirac, sigma=1.5)
G = G / np.sum(G)
# Note: there are also functions to simply apply a gaussian filter
#   e.g. scipy.ndimage.filters.gaussian_filter
# But here the goal is to explicitly create the filter matrix.
# Can you use scipy.ndimage.filters.gaussian_filter to compute the filter matrix?
#  Hint: Which signal do you need to convolve with a gaussian to have exactly the filter as a result?

# compute gradient magnitude and edge indicator function
# for original scale
[dx, dy] = np.gradient(I)
gradMag = np.sqrt(dx**2+dy**2)
g = 1 / (1 + (dx**2+dy**2)/(gamma**2))

# show original scale
plt.figure(figsize=(width, height))
plt.subplot(3, N+1, 1)
plt.imshow(I, cmap='gray')
plt.title('original scale')

plt.subplot(3, N+1, N+1+1)
plt.imshow(gradMag, cmap='gray')
plt.title('gradient magnitude')

plt.subplot(3, N+1, 2*(N+1)+1)
plt.imshow(g)
plt.title('edge indicator')

# filtering, computation of gradient magnitude
# as well as edge indicator function
for i in range(2, N+2):
    # filter image
    I = signal.convolve2d(I, G, mode="same", boundary="symm")

    # recompute gamma (ad hoc choice)
    gamma = gamma/1.1

    # compute gradient magnitude and edge indicator function
    # for this scale
    dx, dy = np.gradient(I)
    gradMag = np.sqrt(dx**2+dy**2)
    g = 1 / (1 + (dx**2+dy**2)/(gamma**2))

    # visualize
    plt.subplot(3, N+1, i)
    plt.imshow(I, cmap='gray')
    plt.title('scale ' + str(i-1))
    plt.subplot(3, N+1, (N+1)+i)
    plt.imshow(gradMag, cmap='gray')
    plt.title('gradient magnitude')
    plt.subplot(3, N+1, 2*(N+1)+i)
    plt.imshow(g)
    plt.title('edge indicator')

plt.show()
```
